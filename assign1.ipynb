### 1. Program to compute the factorial of a given number

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")
```

### 2. Program to display the multiplication table for a given number

```python
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

# Example usage
number = 7
display_multiplication_table(number)
```

### 3. Program to identify if a given number is odd or even

```python
def check_odd_even(num):
    if num % 2 == 0:
        return f"{num} is an even number."
    else:
        return f"{num} is an odd number."

# Example usage
number = 4
print(check_odd_even(number))

number = 7
print(check_odd_even(number))
```

### 4. Program to identify if the entered number is a palindrome or not

```python
def is_palindrome(num):
    # Convert number to string to easily reverse it
    s_num = str(num)
    if s_num == s_num[::-1]:
        return f"{num} is a palindrome."
    else:
        return f"{num} is not a palindrome."

# Example usage
number = 121
print(is_palindrome(number))

number = 123
print(is_palindrome(number))

number = 1221
print(is_palindrome(number))
```

### 5. Develop an Expert System that provides simple decision-making

This is a simple rule-based expert system for animal identification based on a few characteristics.

```python
def animal_expert_system():
    print("Welcome to the Simple Animal Identification Expert System!")
    print("Please answer the following questions with 'yes' or 'no'.")

    # Question 1
    flies = input("Can the animal fly? (yes/no): ").lower()
    if flies == 'yes':
        swims = input("Does the animal swim? (yes/no): ").lower()
        if swims == 'yes':
            print("It could be a duck or a goose.")
        else:
            print("It could be a bird like an eagle or a sparrow.")
    elif flies == 'no':
        has_fur = input("Does the animal have fur? (yes/no): ").lower()
        if has_fur == 'yes':
            eats_meat = input("Does the animal eat meat? (yes/no): ").lower()
            if eats_meat == 'yes':
                print("It could be a lion or a wolf.")
            else:
                print("It could be a cow or a rabbit.")
        else:
            print("It could be a snake or a fish (if it swims).")
    else:
        print("I'm not sure about that characteristic. Please answer with 'yes' or 'no'.")

# Example usage
animal_expert_system()
```
