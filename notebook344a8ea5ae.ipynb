```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=100, freq='D'),
    'Product': np.random.choice(['Laptop', 'Phone', 'Tablet', 'Headphones', 'Shoes'], 100),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 100),
    'Units_Sold': np.random.randint(1, 10, 100),
    'Price': np.random.randint(100, 1000, 100)
})

# Create Sales column
data['Sales'] = data['Units_Sold'] * data['Price']
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data['Month'] = data['Date'].dt.to_period('M')
```

```python
total_sales = data['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
total_units = data['Units_Sold'].sum()
print("Total Units Sold:", total_units)
```

```python
top_products = data.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top_products.plot(kind='bar', color='skyblue')
plt.title("Top 5 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
region_sales = data.groupby('Region')['Sales'].sum()

plt.figure(figsize=(6,6))
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Region")
plt.ylabel('')
plt.show()
```

```python
monthly_sales = data.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line', marker='o', color='green')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.show()
```
