Let's simulate a coin toss experiment multiple times and then compute the probabilities of getting heads or tails.

```python
import numpy as np

# Number of coin tosses
num_tosses = 1000

# Simulate coin tosses (0 for tails, 1 for heads)
# np.random.randint(0, 2, num_tosses) generates random integers between 0 (inclusive) and 2 (exclusive)
coin_tosses = np.random.randint(0, 2, num_tosses)

print(f"First 10 coin tosses: {coin_tosses[:10]}")
```

Now, let's compute the probabilities of heads and tails based on our simulation.

```python
# Calculate the number of heads (where coin_tosses == 1)
num_heads = np.sum(coin_tosses == 1)
# Calculate the number of tails (where coin_tosses == 0)
num_tails = np.sum(coin_tosses == 0)

# Compute probabilities
probability_heads = num_heads / num_tosses
probability_tails = num_tails / num_tosses

# Display the results
print(f"Number of tosses: {num_tosses}")
print(f"Number of heads: {num_heads}")
print(f"Number of tails: {num_tails}")
print(f"Probability of Heads: {probability_heads:.4f}")
print(f"Probability of Tails: {probability_tails:.4f}")
```

Let's generate data from a normal distribution and visualize it with a histogram. We'll use `numpy` for data generation and `matplotlib.pyplot` for plotting.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
sample_size = 1000 # number of data points

# Generate data from a normal distribution
normal_data = np.random.normal(mu, sigma, sample_size)

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram of a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

First, let's create a sample dataset using `numpy`.

```python
import numpy as np

# Generate a sample dataset
data = np.array([12, 15, 18, 20, 22, 25, 28, 30, 32, 35, 40, 45, 50, 55, 60])

print("Sample Dataset:", data)
```

Now, let's compute the mean, median, variance, and standard deviation of this dataset using `numpy`.

```python
import numpy as np

# Compute statistics
mean_value = np.mean(data)
median_value = np.median(data)
variance_value = np.var(data)
std_dev_value = np.std(data)

# Display the results
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value}")
```
